```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/louis/OneDrive/Documents/University of Manchester/Physics/Year 4/MPhys/data/refill/"
```

```python
spark_line_refill = np.genfromtxt(path+'spark_line_refill.csv', delimiter=',')
optical_line_refill = np.genfromtxt(path+'optical_line_refill.csv', delimiter=',')
```

```python
plt.title("Spark Line Refilling")
plt.ylabel("Pressure (mBar)")
plt.xlabel("Time (s)")
#plt.yscale("log")
#plt.xscale("log")
plt.plot(spark_line_refill[:,0], spark_line_refill[:,1], "k.", label='Spark Line')
```

```python
plt.title("Optical Line Refilling")
plt.ylabel("Pressure (mBar)")
plt.xlabel("Time (s)")
#plt.yscale("log")
#plt.xscale("log")
plt.plot(optical_line_refill[:,0], optical_line_refill[:,1], "k.", label='optical Line')
```
