```python
def plot_with_fit(x, y, color, label):
    # Fit a 3rd order polynomial
    coeffs = np.polyfit(x, y, 3)
    poly = np.poly1d(coeffs)
    x_sorted = np.linspace(min(x), max(x), 500)
    y_fit = poly(x_sorted)
    
    # Plot original data
    plt.plot(x, y, color + ".", label=label)
    # Plot fitted polynomial
    plt.plot(x_sorted, y_fit, color + "-")

# Main plot
plt.figure(figsize=(10, 6))
plt.title("Optical Line Refilling", weight = "bold")
plt.ylabel("Pressure (mbar)")
plt.xlabel("Time (s)")

# Get the first 24 elements from column 0
x_vals = refill_25_02_25[:, 0][0:24]
y_vals = refill_25_02_25[:, 3][0:24]

# Create a mask to exclude values equal to 1.65e-05
mask = x_vals != 1.65e-05

# Apply the mask to get filtered arrays
y_filtered = x_vals[mask]
x_filtered = y_vals[mask]


plot_with_fit(optical_line_refill[:, 0], optical_line_refill[:, 1], "g", "13/02/25 IG")
plot_with_fit(refill_18_02_25[:, 0], refill_18_02_25[:, 1], "r", "18/02/25 IG")
plot_with_fit(x_filtered, y_filtered, "c", "25/02/25 IG")
plot_with_fit(optical_20_03_25[:, 0], optical_20_03_25[:, 1], "k", "20/03/25 IG")
plot_with_fit(optical_20_03_25[:, 0], optical_20_03_25[:, 2], "b", "20/03/25 CG")
plot_with_fit(refill_01_04_25[:, 0], refill_01_04_25[:, 1], "m", "01/04/25 IG")
plot_with_fit(refill_01_04_25[:, 0], refill_01_04_25[:, 2], "y", "01/04/25 CG")

# Add legend and ticks
plt.legend()
plt.grid(True)
plt.minorticks_on()
plt.tick_params(axis='both', which='both', direction='in', top=True, right=True)

plt.show()
```
