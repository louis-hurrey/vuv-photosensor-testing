```python
warmup = io.load_rwf_info("C:/Users/louis/OneDrive/Documents/University of Manchester/Physics/Year 4/MPhys/data/warmup_data/warmup_dig1-DT5740-915_CH28_20250501105347-03.h5", samples = 501)
```

```python
warmup_avg = np.zeros(len(warmup.rwf))

for i in range(len(warmup.rwf)):
    warmup_avg[int(i)] = np.mean(warmup.rwf[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate time in hours
time_secs = np.linspace(0, len(warmup.rwf) * 0.5, len(warmup.rwf))
time_hrs = time_secs / 3600

# Plot styling
plt.style.use('default')  # Try 'seaborn-darkgrid' or 'ggplot' for variations
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.plot(time_hrs, warmup_avg, color='tab:blue', linewidth=2)

# Enhance aesthetics
ax.set_title("Average Trigger Baseline as Flash Lamp Warms Up", fontsize=16, weight='bold')
ax.set_xlabel("Time (hours)", fontsize=14)
ax.set_ylabel("Average Trigger Baseline (mV)", fontsize=14)
ax.tick_params(axis='both', labelsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Optional: tighten layout and export
plt.tight_layout()
# plt.savefig("flash_lamp_warmup.png", dpi=300)

plt.show()
```
